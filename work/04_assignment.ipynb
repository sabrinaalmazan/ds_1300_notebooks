```python
import pandas as pd
```

```python
facts = pd.read_csv('../data/state_facts.tsv',delimiter="\t")
facts
```

```python
#lets see if there is connection between gdp and being close to body of water. historically access to water = success
topwat = facts.sort_values(by=['Area_water'], ascending=False).head(5)
lowwat = facts.sort_values(by=['Area_water'], ascending=False).tail(5)
frames = [topwat, lowwat]
joined = pd.concat(frames)
joined
```

```python
plotted2 = joined.plot(kind='scatter', x = 'Area_water', y='GDP_2020', title = 'Water & GDP')
plotted2
```
