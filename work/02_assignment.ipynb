```python
import pandas as pd
```

```python
facts = pd.read_csv('../data/state_facts.tsv',delimiter="\t")
facts.head(5)
```

```python
#question1
facts['Pop_2020_density']= facts['Pop_2020']/facts['Area_land']
fact2020pop= facts.sort_values(by='Pop_2020_density', ascending=False)
fact2020pop.head(5)
```

```python
#question2 
facts.query('Area_water > Area_land')
```

```python
#question3
#create column with pop 2020 - 2010 and sort by ascending
facts['10_Year_Growth_%'] = ((facts['Pop_2020']-facts['Pop_2010']) / (facts['Pop_2020']))*100
factsFastPopGrowth = facts.sort_values(by='10_Year_Growth_%', ascending=False)
factsFastPopGrowth
```

```python
#question 4 population
facts4 = facts.sort_values(by="Pop_2020", ascending = False)
facts4.head(2)
```

```python
#question4landarea
facts4 = facts.sort_values(by='Area_land',ascending=False)
facts4.head(2)
```

```python
#question5
facts.loc[(facts['Capital'] == facts ['City_1'])]
```

```python
#question 6 Which city has the largest percent drop from their largest city to their 5th largest? 100*(1st largest - 5th largest)/(1st largest)
facts['percDrop'] = 100*(facts['city_1_pop']- facts ['city_5_pop'])/(facts['city_1_pop'])
facts.sort_values(by=['percDrop'], ascending=False).head(1)

```
